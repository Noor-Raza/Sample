```python
import matplotlib.pyplot as plt

set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary with all datasets
    df = datasets[dataset]  # Set df as the dataframe

    df['Date'] = pd.to_datetime(df['Date'])  # Convert Date column to datetime format
    df_2023 = df[df['Date'].dt.year == 2023]  # Select only data from 2023

    rows, _ = df_2023.shape  # Extract the row numbers

    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe dataset {dataset} has {rows} rows in the year 2023.\nThe close price trend for 2023 is the following:')

    plt.figure(figsize=(15, 5))  # Set figure size
    plt.plot(df_2023['Date'], df_2023['Close'], color=colors[dataset])  # Plot the Close price by day as line graph
    plt.title(f'{dataset} Close Price Trend (2023)', fontweight='bold')  # Set graph title with bold font
    plt.xlabel('Date')  # Set graph x axis label
    plt.ylabel('Close Price')  # Set graph y axis label
    plt.grid(alpha=0.3)  # Add grid for better readability
    plt.show()  # Display graph
```

### **Exercise 3:**

For each company, find the day with the highest Close price and display it alongside the price.

#### Option 1

```python
from datetime import datetime

# List to store results
result = []

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as the dataframe
    
    max_close_value = df['Close'].max()  # Extract the highest Close price
    max_close_row = df[datasets[dataset]['Close'] == max_close_value]  # Extract the row corresponding to the highest Close price
    max_close_date = max_close_row['Date'].iloc[0].date()  # Extract the date corresponding to the highest Close price
    
    formatted_max_close_date = datetime.strptime(str(max_close_date), '%Y-%m-%d').strftime('%d. %B %Y')  # Format the resulting date for improved readability

    # Append the results to the list
    result.append([dataset, f"{max_close_value:.2f}", formatted_max_close_date])

# Convert the result list to a pandas DataFrame
result_df = pd.DataFrame(result, columns=['Dataset', 'Highest Close Price', 'Date'])

result_df
```

Group the data by month and calculate the average Close price for each company. Plot these monthly averages for 3 companies and compare them. Justify the chart selection.

#### Option 1

```python
import plotly.graph_objects as go

set_counter = 0  # Initialize counter for number of sets
all_avg_close_month = {}  # Initialize dictionary for all companies monthly average Close prices

# Loop through datasets and calculate monthly averages
for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset].copy()  # Set df as copy of the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    avg_close_month = df.groupby(df['Month'])['Close'].mean()  # Calculate the average Close price by month

    all_avg_close_month[dataset] = avg_close_month  # Combine monthly averages into previously initialized dictionary

# Convert dictionary into DataFrame
all_avg_close_month_df = pd.DataFrame(all_avg_close_month).reset_index()  # Reset index to include 'Month'

# Convert the 'Month' column from Period to string
all_avg_close_month_df['Month'] = all_avg_close_month_df['Month'].astype(str)

# Create the interactive table using Plotly
fig = go.Figure(data=[go.Table(
    header=dict(values=list(all_avg_close_month_df.columns),
                fill_color='paleturquoise',
                align='left',
                font=dict(size=12)),
    cells=dict(values=[all_avg_close_month_df[col].tolist() for col in all_avg_close_month_df.columns],
               fill_color='lavender',
               align='left',
               font=dict(size=10))
)])

# Show the table
fig.show()

```

### **Exercise 6:**

For each company create a plot showing the range of prices for each month and justify the chart selection.

#### Graph: Option 1

```python
set_counter = 0  # Initialize counter for number of sets
ranges_month = {}  # Initialize dictionary for each companies monthly Close prices
all_ranges_month = {}  # Initialize dictionary for all companies monthly Close prices

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset].copy()  # Set df as copy of the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    close_max = df.groupby(df['Month'])['Close'].max()  # Calculate the maximum Close price by month
    close_min = df.groupby(df['Month'])['Close'].min()  # Calculate the minimum Close price by month

    ranges_month['Highs'] = close_max  # Enter monthly maximums into previously initialized dictionary
    ranges_month['Lows'] = close_min  # Enter monthly minimums into previously initialized dictionary
    ranges_month_df = pd.DataFrame(ranges_month)  # Convert dictionary into DataFrame

    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe dataset {dataset} range of Close prices is the following:')

    ranges_month_df.plot(figsize=(15, 5))  # Plot the Close prices by month as line graph
    plt.title(f'Range of Monthly Close Prices', fontweight='bold')  # Set graph title
    plt.xlabel('Month')  # Set graph x axis label
    plt.ylabel('Close Prices')  # Set graph y axis label
    plt.grid(alpha=0.3)  # Add grid for better readability
    plt.show()  # Display graph
```

#### **Justification of Chart Selection:**
We chose line charts to display the monthly range of Close prices for each company, as this visualization is well-suited for representing the data's key characteristics.
- **Shows Variations Clearly:** A line chart effectively captures the highs and lows of Close prices for each month, highlighting the range and providing a clear view of fluctuations within the time frame.
- **Facilitates Range Visualization:** By plotting the maximum and minimum Close prices for each month on the same chart, we can directly observe the spread and variability in the prices. This approach makes it easy to identify months with high volatility or stability.

#### Graph: Option 2

```python
set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset].copy()  # Set df as copy of the dataframe

    df['Month'] = df['Date'].dt.to_period('M')  # Add a new column for the month
    close_month = df.groupby('Month')['Close'].apply(list)

    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe dataset {dataset} range of Close prices is the following:')

    plt.figure(figsize=(15, 5))  # Set figure size
    plt.boxplot(close_month, tick_labels=close_month.index)  # Plot boxplots with monthly Close price range
    plt.title(f'{dataset} Close Price Ranges by Month', fontweight='bold')  # Set graph title
    plt.xlabel('Month')  # Set graph x axis label
    plt.ylabel('Close Price')  # Set graph y axis label
    plt.grid(alpha=0.3)  # Add grid for better readability
    plt.show()  # Display graph
```

#### **Justification of Chart Selection:**
We chose box plots to display the range of Close prices for each month because this chart type is specifically suited to represent the distribution and variability of data:
- **Represents the Data Range Clearly:** Box plots provide a concise summary of the distribution of Close prices for each month, showing the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values.
- **Highlights Outliers:** This chart type is particularly effective in identifying outliers, which can be critical in stock price analysis to understand unusual fluctuations or anomalies.
- **Facilitates Monthly Comparison:** By showing the monthly ranges side-by-side, box plots allow for easy comparison of variability and central tendency in Close prices across different months.
- **Compact Representation:** Box plots summarize a large amount of data in a compact and visually straightforward way, making them ideal for analyzing the range of prices over multiple months without overwhelming the viewer.
- **Focus on Variability:** The emphasis on the interquartile range (IQR) and median values makes box plots the most effective choice for understanding the volatility and spread of stock prices.

### **Exercise 10:**

For each company:
- Calculate the spread between the High and Low prices for each day
- Compute the average spread for each company and visualize the result in a chart. Justify the chart selection
- Interpret which companies exhibit the largest spreads and why this might happen

#### Option 1

```python
set_counter = 0  # Initialize counter for number of sets

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Spread'] = df['High'] - df['Low']  # Add new column with the Spread between High and Low prices
    set_counter += 1  # Add to the counter for each iteration
    print(f'\n{set_counter}) {dataset}\nThe spread for {dataset} is the following:\n{df["Spread"]}')
```

```python
all_avg_spread = {}  # Initialize dictionary for all companies average Spread
bar_colors = []  # Initialize list of colors for each company

for dataset in datasets:  # Iterate through dictionary
    df = datasets[dataset]  # Set df as copy of the dataframe

    df['Avg Spread'] = df['Spread'] / ((df['High'] + df['Low']) / 2)  # Add new column with the average Spread between the average of High and Low prices
    all_avg_spread[dataset] = df['Avg Spread'].mean()*100  # Combine average Spreads into previously initialized dictionary

    bar_colors.append(colors[dataset])  # Add color to the previously created list

plt.figure(figsize=(15, 5))  # Set figure size
plt.bar(all_avg_spread.keys(), all_avg_spread.values(), color=bar_colors)  # Plot the average Spreads as bar chart
plt.title(f'Comparison of Average Spreads by Company', fontweight='bold')  # Set chart title
plt.xlabel('Company')  # Set chart x axis label
plt.ylabel('Average Spread')  # Set chart y axis label
plt.grid(alpha=0.3)  # Add grid for better readability
plt.show()  # Display chart
```

#### **Justification of Chart Selection:**
We chose a bar chart to visualize the average spread between the High and Low prices for each company because it effectively highlights differences across categories:
- **Comparison Across Companies:** A bar chart is ideal for comparing discrete categories (companies) and their corresponding values (average spread). It provides a clear visual representation of which companies exhibit larger or smaller spreads.
- **Focus on Magnitude:** Bar charts emphasize the magnitude of the spreads, making it easy to identify companies with the largest or smallest average spreads at a glance.
- **Categorical Data Representation:** Since each company represents a distinct category, a bar chart is more appropriate than other chart types (e.g., line charts), which are better suited for continuous data.
- **Insightful Visual:** The bar chart effectively supports the interpretation of which companies exhibit larger spreads and encourages further discussion on the reasons behind this variability (e.g., market volatility, liquidity, or stock price levels).
#### **Understanding from the Chart Selection:**
- **Volatility Patterns:** The variation in spreads across companies might reflect differences in market capitalization, investor demand, or the nature of trading activity in these stocks. Larger spreads may also indicate stocks that are less frequently traded or subject to speculative interest.
- **Market Volatility and Liquidity:** Larger spreads, such as for LKNCY, might suggest either higher market volatility or lower liquidity, where fewer trades lead to greater differences between High and Low prices.
